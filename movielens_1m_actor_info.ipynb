```python
# 筛选包含 NaN 值的行
mask = infos.isna().any(axis=1)
rows_with_nan = infos[mask]
rows_with_nan
```

```python
rows_with_nan
rows_with_nan['name'].values
```

```python
# 定义一个函数，用于去除括号以外的内容
def remove_parentheses(string):
    return re.sub(r'\([^()]*\)', '', string)

# 使用apply方法将函数应用到每一行数据
rows_with_nan['name_without_year'] = rows_with_nan['name'].apply(remove_parentheses)

#rows_with_nan['name_without_year'] = rows_with_nan['name'].str.extract(r'^([\w\s]+)')
# 使用 replace() 方法将 NaN 转换为 None
rows_with_nan.replace(np.nan, None, inplace=True)
rows_with_nan['name_without_year'].values
```
