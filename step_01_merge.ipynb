```python
df.drop(columns=['create_time','completed','update_time','name'],inplace=True)
df.rename(columns={'full_name':'name'},inplace=True)
df
```

```python
infos.merge(df,how='left',on='name')
infos.drop_duplicates(subset='name', keep='last', inplace=True)
infos
```

```python
infos = infos.fillna('None')
```

```python
# 将 df1 和 df2 按照 'name' 列进行合并，保留 df1 中的所有行和 df2 中的匹配行
merged = pd.merge(infos, df, on='name', how='left')

# 根据您的要求，将 df1 中的不合法数字值替换为 df2 中相应行的值
for i, row in merged.iterrows():
    if row['genre_x'] == '' or row['genre_x'] == 'None':
        if not pd.isna(row['genre_y']):
            merged.at[i, 'genre_x'] = row['genre_y']
    if row['directors_x'] == '':
        if not pd.isna(row['directors_y']):
            merged.at[i, 'directors_x'] = row['directors_y']
    if row['writers_x'] == 'None':
        try:
            if not pd.isna(row['writers_y']):
                merged.at[i, 'writers_x'] = row['writers_y']
        except:
            print(row['name'])
    if row['starts_x'] == 'None':
        try:
            if not pd.isna(row['starts_y']):
                merged.at[i, 'starts_x'] = row['starts_y']
        except:
            print(row['name'])

# 移除 '_y' 后缀的列，并将 '_x' 后缀的列重命名为原始列名
merged=merged.drop(columns=['genre_y', 'directors_y', 'writers_y', 'starts_y'], axis=1)
merged=merged.drop(columns=['id_y','url_y','time_y','release_time_y','intro_y'],axis=1)
merged=merged.rename(columns={
    'genre_x': 'genre', 'directors_x': 'directors', 'writers_x': 'writers', 'starts_x': 'starts',
    'id_x':'id','url_x':'url','time_x':'time','release_time_x':'release_time','intro_x':'intro'
})

merged.drop_duplicates(subset='name', keep='last', inplace=True)
merged
```

```python
# 筛选包含 NaN 值的行
mask = merged.isna().any(axis=1)
rows_with_nan = merged[mask]
rows_with_nan
```

```python
merged.to_csv(FOLDER+"/merged.csv",index=False)
```
